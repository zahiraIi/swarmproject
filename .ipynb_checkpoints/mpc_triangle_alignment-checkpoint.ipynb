
# Model Predictive Control of 3-Bot Swarm to Achieve Alignment

In this notebook, we formulate and solve a model predictive control (MPC) problem to keep a swarm of three interacting bots aligned in a straight line. The bots experience pairwise interactions, and their motion is governed by nonlinear overdamped dynamics. We use MPC to control their angular velocities \( \omega_i(t) \) such that the area of the triangle they form approaches zero.

### Objective
Minimize the misalignment cost:

$J = \sum_{t=1}^{T} A(\mathbf{X}_t)^2 + \lambda \sum_{t=1}^{T-1} \| \mathbf{u}_t - \mathbf{u}_{t-1} \|^2$

Where:
- $A(\mathbf{X}_t)$ is the area of the triangle formed by the three bots at time \( t \)
- $\mathbf{u}_t = [\omega_1(t), \omega_2(t), \omega_3(t)]^T$ are the spinning frequencies (controls)
- $\lambda$ is a regularization term to encourage smooth control inputs

We use SciPy's `minimize` function with the L-BFGS-B algorithm to solve the optimization problem.


```python
#Imports and Parameters
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# Physical and simulation parameters
alpha = 0.7
beta = 0.7
R0 = 0.5
f0 = 0.05
dt = 0.05
T = 10.0
Nt = int(T / dt)  # Number of steps
N_BOTS = 3

# Initial configuration: aligned along x-axis
a = 2.0
X0 = np.array([[-a, a], [0, 0], [a, a]]).reshape(-1)  # Shape: (6,)
omega_0 = np.array([1,1,1])
```

## Force and Dynamics Definitions
Each bot experiences forces due to other bots based on a radial and tangential interaction model:

- **Radial force**:
\begin{equation}
F_r = \frac{\alpha \omega_i}{r} - \frac{\beta}{(r - R_0)^6 + \epsilon}
\end{equation}

- **Tangential force**:
\begin{equation}
F_t = \frac{f_0 \omega_i}{r}
\end{equation}

The net force is computed for each bot and integrated over time using a 4th-order Runge-Kutta method.



```python
def compute_forces(X, u):
    pos = X.reshape(N_BOTS, 2)
    forces = np.zeros_like(pos)
    for i in range(N_BOTS):
        for j in range(N_BOTS):
            if i == j:
                continue
            rij = pos[j] - pos[i]
            dist = np.linalg.norm(rij) + 1e-6
            rhat = rij / dist
            Fr = alpha * u[i] / dist - beta / ((dist - R0)**6 + 1e-3)
            perp = np.array([-rhat[1], rhat[0]])
            Ft = f0 * u[i] / dist
            forces[i] += Fr * rhat + Ft * perp
    return forces.reshape(-1)

def rk4_step(X, u):
    k1 = compute_forces(X, u)
    k2 = compute_forces(X + 0.5 * dt * k1, u)
    k3 = compute_forces(X + 0.5 * dt * k2, u)
    k4 = compute_forces(X + dt * k3, u)
    return X + dt * (k1 + 2*k2 + 2*k3 + k4) / 6
```

## Area and Cost Function 
To enforce alignment, we minimize the area of the triangle formed by the three bots:

$A = \frac{1}{2} | (x_2 - x_1)(y_3 - y_1) - (x_3 - x_1)(y_2 - y_1) |$

This expression becomes zero when the bots are collinear.


```python
def area(X):
    p = X.reshape(N_BOTS, 2)
    x1, y1 = p[0]
    x2, y2 = p[1]
    x3, y3 = p[2]
    return 0.5 * abs((x2 - x1)*(y3 - y1) - (x3 - x1)*(y2 - y1))

# def objective(U_flat):
#     U = U_flat.reshape(Nt, N_BOTS)
#     X = X0.copy()
#     cost = 0.0
#     for t in range(Nt):
#         u = U[t]
#         X = rk4_step(X, u)
#         cost += area(X)**2
#         if t > 0:
#             du = U[t] - U[t - 1]
#             cost += 0.1 * np.sum(du**2)  # smoothness penalty
#     return cost

def objective(U_flat):
    # U = U_flat.reshape(Nt, N_BOTS)
    U_rest = U_flat.reshape(Nt - 1, N_BOTS)
    U = np.vstack([omega_0, U_rest])  # (N, 3)
    penalty = 1

    X = X0.copy()
    cost = 0.0
    for t in range(Nt):
        u = U[t]
        X = rk4_step(X, u)
        cost += area(X)**2
        if t > 0:  # smoothness: only for t â‰¥ 2 because U[0] is fixed
            du = U[t] - U[t - 1]
            cost += penalty * np.sum(du**2)
    return cost
```

```python
# Optimization Step

# U0 = np.zeros((Nt, N_BOTS)).flatten()
# result = minimize(objective, U0, method='L-BFGS-B', options={'disp': True, 'maxiter': 200})
# U_opt = result.x.reshape(Nt, N_BOTS)

U0 = np.zeros((Nt - 1, N_BOTS)).flatten()
result = minimize(objective, U0, method='L-BFGS-B', options={'disp': True, 'maxiter': 200})
U_opt = result.x.reshape(Nt - 1, N_BOTS)
U_full = np.vstack([omega_0, U_opt])  # final full control sequence
```

```python
#Forward Simulation with Optimal Control
X = X0.copy()
trajectory = [X.reshape(N_BOTS, 2)]
for t in range(Nt):
    X = rk4_step(X, U_full[t])
    trajectory.append(X.reshape(N_BOTS, 2))
trajectory = np.array(trajectory)
```
